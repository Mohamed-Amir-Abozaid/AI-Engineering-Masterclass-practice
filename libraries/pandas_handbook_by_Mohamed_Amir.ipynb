# Core Data Structures in Pandas

Pandas is built on two main data structures:  
1. **Series** → One-dimensional (like a single column in Excel)  
2. **DataFrame** → Two-dimensional (like a full spreadsheet or SQL table)  

---

## Series — 1D Labeled Array
A **Series** is like a list with labels (index).

```python
import pandas as pd

s = pd.Series([10, 20, 30, 40])
print(s)
````

**Output:**

```
0    10
1    20
2    30
3    40
dtype: int64
```

Notice the automatic index: `0, 1, 2, 3`. You can also define a custom index:

```python
s = pd.Series([10, 20, 30], index=["a", "b", "c"])
```

> A Pandas Series may look similar to a Python dictionary because both store data with labels, but a Series offers much more:
>
> * Fast vectorized operations
> * Automatic index alignment during arithmetic
> * Handles missing data using `NaN`
> * Allows label-based and position-based access
> * Integrates seamlessly with Pandas DataFrames

---

## DataFrame — 2D Labeled Table

A **DataFrame** is like a dictionary of Series — multiple columns with labels.

```python
data = {
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["Delhi", "Mumbai", "Bangalore"]
}

df = pd.DataFrame(data)
print(df)
```

**Output:**

```
      name  age      city
0    Alice   25     Delhi
1      Bob   30    Mumbai
2  Charlie   35  Bangalore
```

* Each column in a DataFrame is a **Series**.
* Every Series and DataFrame has an **Index** — helps with:

  * Fast lookups
  * Aligning data
  * Merging & joining
  * Time series operations

```python
df.index     # Row labels
df.columns   # Column labels
```

* You can change them:

```python
df.index = ["a", "b", "c"]
df.columns = ["Name", "Age", "City"]
```

## Data Selection & Filtering

### Selecting Columns

```python
df["column_name"]       # Single column (Series)
df[["col1", "col2"]]   # Multiple columns (DataFrame)
```

### Selecting Rows

```python
df.loc[0]    # First row by label
df.iloc[0]   # First row by position
```

### Specific Rows & Columns

```python
df.loc[0, "Name"]
df.iloc[0, 1]
df.loc[0:2, ["Name", "Age"]]

```

### Fast Single Element Access


### Filtering with Conditions


## Data Cleaning & Preprocessing

### Handling Missing Values

```python
df.isnull()
df.isnull().sum()
df.dropna()
df.fillna(0)
df["Age"].fillna(df["Age"].mean())
df.ffill()
df.bfill()
```

### Detecting Duplicates

```python
df.duplicated()
df.drop_duplicates()
df.duplicated(subset=["Name", "Age"])
```

### String Operations

```python
df["Name"].str.lower()
df["City"].str.contains("delhi", case=False)
df["Email"].str.split("@")
```

### Type Conversion

```python
df["Age"] = df["Age"].astype(int)
df["Date"] = pd.to_datetime(df["Date"])
df["Category"] = df["Category"].astype("category")
```

### Apply Functions

```python
df["Age Group"] = df["Age"].apply(lambda x: "Adult" if x >= 18 else "Minor")
df["Gender"] = df["Gender"].map({"M": "Male", "F": "Female"})
df["City"].replace({"Del": "Delhi", "Mum": "Mumbai"})
```

### Sorting & Ranking

```python
df.sort_values("Age")

df.reset_index(drop=True, inplace=True)

df["Rank"] = df["Score"].rank(method="dense")
```


#### Explain what happens to the keys when a Series is converted to a dictionary via to_dict().

When you convert a Series → Dictionary using to_dict(), this happens exactly:

Each label (index) in the Series becomes a key in the dictionary.

Each value in the Series becomes the value in the dictionary.

The result is a regular dictionary, just like any dictionary in Python.

✅ In short: the Series’ index becomes the dictionary keys, and the Series’

array

## المعنى العام

### 🔹 `map`

> تحويل **Series واحدة** عنصر عنصر

### 🔹 `apply`

> تشغيل **function** على:

* Series
* أو DataFrame (rows / columns)

---

## `map` 🗺️

### تستخدم إمتى؟

✔ لما:

* شغال على **Series فقط**
* كل عنصر لوحده
* transformation بسيط

### مثال:

```python
s = pd.Series([1, 2, 3])

s.map(lambda x: x * 2)
```

**الناتج:**

```
2  4  6
```

### كمان `map` ينفع مع dict

```python
s.map({1: 'A', 2: 'B', 3: 'C'})
```

---

## `apply` 🔧

### تستخدم إمتى؟

✔ لما:

* محتاج logic أعقد
* أو شغال على **DataFrame**
* أو محتاج تتعامل مع row كاملة

---

### على Series

```python
s.apply(lambda x: x * 2)
```

نفس نتيجة `map`

---

### على DataFrame

```python
df.apply(sum)        # column-wise (default)
df.apply(sum, axis=1) # row-wise
```

---

## فرق مهم جدًا ⚠️

### `apply` مع DataFrame

* `axis=0` → أعمدة
* `axis=1` → صفوف

---

## مقارنة سريعة 🔥

| المقارنة            | map | apply |
| ------------------- | --- | ----- |
| يشتغل على Series    | ✔   | ✔     |
| يشتغل على DataFrame | ❌   | ✔     |
| يقبل dict           | ✔   | ❌     |
| أسرع                | ✔   | ❌     |
| Logic معقد          | ❌   | ✔     |

---

## إمتى أستخدم إيه؟

* Series بسيطة → `map`
* استبدال قيم → `map`
* Row logic / conditions → `apply`
* DataFrame → `apply`

---

## خطأ شائع ❌

 **deep copy vs shallow copy**

---

## الأول: المشكلة أصلًا

في Pandas لما تكتب:

```python
kl_head = kl.head()
kl_head[0] = 999
```

Pandas يقولك:

> هو ده View ولا Copy؟ 🤔
> فـ يطلع **SettingWithCopyWarning**

---

## Shallow Copy (نسخة سطحية)

```python
kl_copy = kl.head()
```

### بيعمل إيه؟

* ممكن يشاور على **نفس البيانات في الذاكرة**
* أي تعديل ممكن:

  * يأثر على `kl`
  * أو لا
* سلوك **غير مضمون** ❌

---

## Deep Copy (نسخة عميقة) ✅

```python
kl_copy = kl.head().copy(deep=True)
```

### بيعمل إيه؟

* ينسخ **البيانات نفسها**
* ينسخ الـ index
* أي تعديل:

  * **مش هيأثر** على `kl`
  * ومفيش warnings 🎉

---

## الفرق بالمثال 🔥

```python
kl = pd.Series([10, 20, 30, 40, 50, 60])

# shallow
a = kl.head()
a.iloc[0] = 999

# deep
b = kl.head().copy()
b.iloc[0] = 888
```

### النتيجة:

* `kl` ممكن يتغيّر بسبب `a` ❌
* `kl` **مش** هيتغيّر بسبب `b` ✔

---

## ليه SettingWithCopyWarning مهم؟

Pandas بيحذّرك:

> "أنا مش متأكد إنت بتعدّل في Copy ولا View"

والحل الصح دايمًا:

```python
.copy()
```

---

## مقارنة سريعة 🧠

| المقارنة        | Shallow | Deep   |
| --------------- | ------- | ------ |
| نفس الذاكرة     | ممكن    | ❌      |
| آمن للتعديل     | ❌       | ✔      |
| SettingWithCopy | ممكن    | ❌      |
| استخدامه        | نادر    | الأفضل |

---

## الخلاصة الذهب 🧠

> **عايز تعدّل؟ اعمل Deep Copy**
> `copy(deep=True)` = راحة بال 😎

الكود بتاعك:

```python
kl_copy = kl.head().copy(deep=True)
```

✅ احترافي
✅ بدون warnings
