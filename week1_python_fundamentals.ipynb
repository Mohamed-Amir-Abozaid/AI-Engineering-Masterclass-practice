````python
# ML Example: Accuracy function
def calculate_accuracy(predictions, true_labels):
    """
    Calculate classification accuracy.

    Args:
        predictions: List of predicted labels
        true_labels: List of true labels

    Returns:
        Accuracy as percentage (0-100)
    """
    correct = 0
    for i in range(len(predictions)):
        if predictions[i] == true_labels[i]:
            correct += 1  # this is the same as correct = correct + 1

    accuracy = (correct / len(predictions)) * 100
    return accuracy


# Test it
pred = [0, 1, 1, 0, 1]
true = [0, 1, 0, 0, 1]
acc = calculate_accuracy(pred, true)
print(f"Accuracy: {acc:.1f}%")
````

````python
# Euclidean distance
def distance(point1, point2):
    """
    Calculate distance between two 2D points.

    Formula: sqrt((x1-x2)Â² + (y1-y2)Â²)
    """
    x1, y1 = point1
    x2, y2 = point2

    dist = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return dist


# Test
p1 = [0, 0]
p2 = [3, 4]
d = distance(p1, p2)
print(f"Distance from {p1} to {p2}: {d}")
````

````python
# Normalization function (preview this will be discussed in the lecture in detail but see how to implement it here!)
def normalize(values):
    """
    Normalize values to [0, 1] range.


    """
    min_val = min(values)
    max_val = max(values)
    range_val = max_val - min_val

    normalized = []
    for val in values:
        norm_val = (val - min_val) / range_val
        normalized.append(norm_val)

    return normalized


# Test
data = [10, 20, 30, 40, 50]
norm_data = normalize(data)
print(f"Original: {data}")
print(f"Normalized: {[f'{x:.2f}' for x in norm_data]}")
````

### Exercise 4.3: Normalization Function


````python
def normalize_list(values):
    """
    Normalize to [0,1] using min-max scaling.

    Formula: (x - min) / (max - min)
    """
    # YOUR CODE HERE
    #raise NotImplementedError()
    normalized = []
    for val in values:
      norm = val - min(values)
      norm /= max(values) - min(values)
      normalized.append(norm)
    return normalized


# Test
data = [10, 20, 30, 40, 50]
result = normalize_list(data)
print(f"Normalized: {[f'{x:.2f}' for x in result]}")
````
